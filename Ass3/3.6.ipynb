```python
The diagram represents how a simple neuron works:

* It takes inputs (**X1, X2, X3**).
* Each input has a weight (**w1, w2, w3**).
* A bias (**b**) is added.
* These are combined into a **weighted sum**.
* The sum is passed through an **activation function** (in this case, the hyperbolic tangent `tanh`).
* The final output is produced.

---

### ðŸ“Œ Process in words

1. Multiply each input by its corresponding weight.
2. Add them all together.
3. Add the bias.
4. Pass the result into the activation function `tanh`.

---

### ðŸ§® Example with given values

* Inputs:

  * X1 = 2.0
  * X2 = 3.0
  * X3 = 4.0
* Weights:

  * w1 = 0.05
  * w2 = 0.07
  * w3 = 0.09
* Bias:

  * b = 0.60

---

### ðŸ”¢ Step-by-step calculation

1. Multiply each input by its weight:

   * X1 Ã— w1 = 2.0 Ã— 0.05 = 0.10
   * X2 Ã— w2 = 3.0 Ã— 0.07 = 0.21
   * X3 Ã— w3 = 4.0 Ã— 0.09 = 0.36

2. Add them together:

   * 0.10 + 0.21 + 0.36 = 0.67

3. Add the bias:

   * 0.67 + 0.60 = 1.27

4. Apply the activation function `tanh`:

   * tanh(1.27) â‰ˆ 0.8538

---

### âœ… Final Result

* The **weighted sum** = 1.27
* The **output after tanh** activation = 0.8538

---

ðŸ‘‰ In simple words: the neuron takes the inputs, weights them, adds a bias, and then squashes the result into a value between -1 and 1 using the `tanh` function.
```

```python
import numpy as np

X = np.array([2.0, 3.0, 4.0])

W = np.array([0.05, 0.07, 0.09])

b = 0.60

weighted_sum = np.dot(X, W) + b
print("Weighted Sum =", weighted_sum)

output = np.tanh(weighted_sum)
print("Output (tanh) =", output)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Inputs and weights (from the diagram)
X = np.array([2.0, 3.0, 4.0])
W = np.array([0.05, 0.07, 0.09])
b = 0.60

# Weighted sum
def neuron_sum(X, W, b):
    return np.dot(X, W) + b

# Activation functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

def tanh(x):
    return np.tanh(x)

# Compute weighted sum
z = neuron_sum(X, W, b)

# Apply activations
output_sigmoid = sigmoid(z)
output_relu = relu(z)
output_tanh = tanh(z)

print("Weighted Sum (z) =", z)
print("Sigmoid Output   =", output_sigmoid)
print("ReLU Output      =", output_relu)
print("Tanh Output      =", output_tanh)

# Visualization
x_vals = np.linspace(-3, 3, 200)

plt.figure(figsize=(12, 4))

# Sigmoid
plt.subplot(1, 3, 1)
plt.plot(x_vals, sigmoid(x_vals), label="Sigmoid")
plt.axvline(z, color='red', linestyle='--')
plt.scatter(z, output_sigmoid, color='red')
plt.scatter(z, output_sigmoid, color='red')
plt.title("Sigmoid Activation")
plt.legend()

# ReLU
plt.subplot(1, 3, 2)
plt.plot(x_vals, relu(x_vals), label="ReLU")
plt.axvline(z, color='red', linestyle='--')
plt.scatter(z, output_relu, color='red')
plt.title("ReLU Activation")
plt.legend()

# Tanh
plt.subplot(1, 3, 3)
plt.plot(x_vals, tanh(x_vals), label="Tanh")
plt.axvline(z, color='red', linestyle='--')
plt.scatter(z, output_tanh, color='red')
plt.title("Tanh Activation")
plt.legend()

plt.tight_layout()
plt.show()
```
