```python
import matplotlib.pyplot as plt
import numpy as np

# ===== Dataset =====
students = ["Trang", "Thảo", "Dương", "Dũng", "Sơn", "Phúc", "Quý", "Quang Anh"]
subjects = ["Math", "Physics", "Chemistry", "Biology", "English"]

marks = {
    "Trang":     [85, 78, 92, 74, 88],
    "Thảo":      [72, 69, 80, 75, 70],
    "Dương":     [90, 88, 84, 91, 85],
    "Dũng":      [65, 70, 68, 60, 72],
    "Sơn":       [78, 82, 79, 85, 80],
    "Phúc":      [88, 85, 92, 87, 90],
    "Quý":       [70, 73, 68, 72, 75],
    "Quang Anh": [95, 92, 90, 94, 93]
}

# ===== 1. Multiple Line Chart =====
plt.figure(figsize=(10,6))
for student in students:
    plt.plot(subjects, marks[student])
plt.title("Student Marks (Multiple Lines)")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.show()

# ===== 2. Line Chart with Legend =====
plt.figure(figsize=(10,6))
for student in students:
    plt.plot(subjects, marks[student], marker="o", label=student)
plt.title("Student Marks (With Legend)")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.legend()
plt.show()

# ===== 3. Bar Chart =====
x = np.arange(len(subjects))  # vị trí các môn
width = 0.1  # độ rộng của cột

plt.figure(figsize=(12,6))

for i, student in enumerate(students):
    plt.bar(x + i*width - (len(students)/2)*width, marks[student], width, label=student)

plt.title("Student Marks (Bar Chart)")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.xticks(x, subjects)  # gán tên môn học trên trục x
plt.legend(ncol=4)  # chia legend thành 4 cột cho gọn
plt.grid(True, color="gray", linestyle="--", linewidth=0.5)
plt.show()
```
