```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# 1. Dữ liệu gốc
X = np.array([50, 60, 65, 70, 75, 80, 85]).reshape(-1, 1)  # diện tích
y = np.array([2.5, 3.0, 3.5, 3.8, 4.0, 4.5, 5.0])          # giá nhà

# 2. Khởi tạo và huấn luyện mô hình
model = LinearRegression()
model.fit(X, y)

# 3. Dự đoán giá nhà cho diện tích mới
X_new = np.array([55, 68, 76, 90]).reshape(-1, 1)
y_pred = model.predict(X_new)

# 4. In kết quả
for size, price in zip([55, 68, 76, 90], y_pred):
    print(f"Diện tích {size} m² → Giá dự đoán: {price:.2f} tỷ")

# 5. Vẽ biểu đồ trực quan
plt.scatter(X, y, color="blue", label="Dữ liệu thực tế")
plt.plot(X, model.predict(X), color="red", label="Hồi quy tuyến tính")
plt.scatter(X_new, y_pred, color="green", marker="x", s=100, label="Dự đoán")
plt.xlabel("Diện tích (m²)")
plt.ylabel("Giá nhà (tỷ)")
plt.title("Dự đoán giá nhà theo diện tích")
plt.legend()
plt.show()
```
