```python
import numpy as np

# Input arrays
names = np.array(['Ann', 'Joe', 'Mark'])
heights = np.array([1.5, 1.78, 1.6])   # in meters
weights = np.array([65, 46, 59])       # in kilograms

# Step 1: Calculate BMI = weight / (height^2)
bmi = weights / (heights ** 2)

# Step 2: Classify BMI categories
categories = np.empty(bmi.shape, dtype="object")

categories[bmi < 18.5] = "Underweight"
categories[(bmi >= 18.5) & (bmi <= 25)] = "Normal weight"
categories[bmi > 25] = "Overweight"

# Step 3: Display results
for name, h, w, b, cat in zip(names, heights, weights, bmi, categories):
    print(f"{name}: Height={h}m, Weight={w}kg, BMI={b:.2f}, Category={cat}")
```
