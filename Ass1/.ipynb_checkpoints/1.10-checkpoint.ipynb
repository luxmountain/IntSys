```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Đọc dữ liệu Titanic từ file CSV
df = pd.read_csv("titanic.csv")

# 2. Tạo cột phân loại mới: Person (Men, Women, Children)
def get_person(row):
    if row['age'] < 18:
        return 'Children'
    elif row['sex'] == 'male':
        return 'Men'
    else:
        return 'Women'

df['person'] = df.apply(get_person, axis=1)

# 3. Tính tỉ lệ sống sót theo class và person
survival_rate = df.groupby(["class", "person"])["survived"].mean().reset_index()

print(survival_rate)

# 4. Vẽ biểu đồ
sns.catplot(
    data=survival_rate,
    x="class",
    y="survived",
    hue="person",
    kind="bar",
    height=5,
    aspect=1.2
)

plt.ylabel("Survival Rate")
plt.title("Survival Rate of Men, Women, and Children in Titanic Classes")
plt.show()
```
