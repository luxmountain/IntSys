```python
# ==========================
# Full corrected pipeline (a -> h) - pxquy ‚úÖ Final
# ==========================
%matplotlib inline

import os, random, csv, json, pickle, re, shutil, warnings
from pathlib import Path
from datetime import datetime, timedelta
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix
from sklearn.utils.class_weight import compute_class_weight
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Conv1D, GlobalMaxPooling1D, Dense, Dropout, SimpleRNN, LSTM, Bidirectional, BatchNormalization
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau

warnings.filterwarnings("ignore")
SEED = 42
random.seed(SEED)
np.random.seed(SEED)

# --------------------------
# b) Generate improved dataset
# --------------------------
OUTFILE = "itemReview_pxquy.csv"
print("üîÅ Generating dataset ->", OUTFILE)

NUM_USERS = 1000
NUM_ITEMS = 500
REVIEWS_PER_USER = 50

users = [f"user_{i:04d}" for i in range(1, NUM_USERS+1)]
items = [f"item_{i:04d}" for i in range(1, NUM_ITEMS+1)]

pos_phrases = [
    "S·∫£n ph·∫©m th·∫≠t s·ª± r·∫•t t·ªët","M√¨nh c·∫£m th·∫•y h√†i l√≤ng khi s·ª≠ d·ª•ng",
    "Thi·∫øt k·∫ø ƒë·∫πp, ch·∫•t l∆∞·ª£ng v∆∞·ª£t mong ƒë·ª£i","ƒê√°ng ti·ªÅn, s·∫Ω gi·ªõi thi·ªáu cho b·∫°n b√®",
    "Giao h√†ng nhanh, ƒë√≥ng g√≥i c·∫©n th·∫≠n","Ch·∫•t l∆∞·ª£ng c·ª±c k·ª≥ t·ªët","Tr·∫£i nghi·ªám r·∫•t tuy·ªát v·ªùi"
]
neg_phrases = [
    "S·∫£n ph·∫©m qu√° t·ªá, kh√¥ng ƒë√°ng ti·ªÅn","M√¨nh r·∫•t th·∫•t v·ªçng v·ªÅ ch·∫•t l∆∞·ª£ng",
    "H√†ng nh·∫≠n ƒë∆∞·ª£c kh√¥ng gi·ªëng m√¥ t·∫£","Giao h√†ng ch·∫≠m, ƒë√≥ng g√≥i c·∫©u th·∫£",
    "S·∫Ω kh√¥ng bao gi·ªù mua l·∫°i","Th·ª±c s·ª± l√† m·ªôt tr·∫£i nghi·ªám t·ªá h·∫°i","Kh√¥ng h·ªÅ ƒë√°ng v·ªõi s·ªë ti·ªÅn b·ªè ra"
]
neutral_phrases = [
    "S·∫£n ph·∫©m ·ªü m·ª©c b√¨nh th∆∞·ªùng, kh√¥ng c√≥ g√¨ ƒë·∫∑c bi·ªát","Ch·∫•t l∆∞·ª£ng t·∫°m ·ªïn, gi√° c·∫£ h·ª£p l√Ω",
    "Kh√¥ng qu√° t·ªá nh∆∞ng c≈©ng kh√¥ng qu√° t·ªët","T·∫°m ch·∫•p nh·∫≠n ƒë∆∞·ª£c",
    "Kh√° ·ªïn nh∆∞ng c√≤n nhi·ªÅu ƒëi·ªÉm c·∫ßn c·∫£i thi·ªán"
]
negation_patterns = ["Kh√¥ng {} ch√∫t n√†o","Ho√†n to√†n kh√¥ng {}","Ch∆∞a th·∫•y {} bao gi·ªù","Kh√¥ng h·ªÅ {}"]
mixed_patterns = ["M·∫∑c d√π {}, nh∆∞ng {}","{}. Tuy nhi√™n, {}"]
noise_tokens = ["∆°","√†","·ªù","hmm","ki·ªÉu","vl","ggg","·ªùm","n√≥i chung","ki·ªÉu nh∆∞"]
punct_noise = ["!!!","???","...",":))","üòÇ","@@@","###"]
slang_map = {
    "kh√¥ng":["ko","khong"],"t·ªët":["tot","t0t"],"x·∫•u":["xau"],"ch·∫•t l∆∞·ª£ng":["chat luong","clg"],
    "h√†i l√≤ng":["hai long","hailong"],"ƒë√°ng":["dang"]
}

def inject_slang_typo(text):
    for k, vs in slang_map.items():
        if k in text and random.random() < 0.25:
            text = text.replace(k, random.choice(vs))
    if random.random() < 0.05:
        text = text.replace("ƒë","d").replace("ƒê","D")
    return text

def make_long_review(rating):
    sents = []
    for _ in range(random.randint(3,6)):
        if rating >= 4:
            s = random.choice(pos_phrases)
            if random.random() < 0.18:
                s = random.choice(negation_patterns).format("t·ªët")
        elif rating == 3:
            s = random.choice(neutral_phrases)
        else:
            if random.random() < 0.35:
                s = random.choice(mixed_patterns).format(random.choice(pos_phrases), random.choice(neg_phrases))
            else:
                s = random.choice(neg_phrases)
        s = inject_slang_typo(s)
        if random.random() < 0.42: s += " " + random.choice(noise_tokens)
        if random.random() < 0.35: s += " " + random.choice(punct_noise)
        sents.append(s)
    return ". ".join(sents)

start_date = datetime(2023, 1, 1)
with open(OUTFILE,"w",newline="",encoding="utf-8") as f:
    w = csv.writer(f); w.writerow(["user_id","item_id","rating","review_text","date"])
    for u in users:
        for _ in range(REVIEWS_PER_USER):
            item = random.choice(items)
            p = random.random()
            rating = 1 if p<0.14 else 2 if p<0.30 else 3 if p<0.68 else 4 if p<0.86 else 5
            review = make_long_review(rating)
            if random.random() < 0.12: rating = random.randint(1,5) # label noise
            date = (start_date+timedelta(days=random.randint(0,1000))).strftime("%Y-%m-%d")
            w.writerow([u,item,rating,review,date])
print("‚úÖ Dataset saved:", OUTFILE)

# --------------------------
# c) Preprocess
# --------------------------
df = pd.read_csv(OUTFILE)
def preprocess_text(text):
    text = str(text).lower()
    text = re.sub(r"[^0-9a-z·∫°√°√†·∫£√£ƒÉ·∫Ø·∫±·∫≥·∫µ√¢·∫•·∫ß·∫©·∫´ƒë√™·∫ø·ªÅ·ªÉ·ªÖ√¨√≠·ªâƒ©√≤√≥·ªè√µ√¥·ªë·ªì·ªï·ªó∆°·ªõ·ªù·ªü·ª£√π√∫·ªß≈©∆∞·ª©·ª´·ª≠·ªØ·ª≥√Ω·ª∑·ªπ\s\.\,\!\?]", " ", text)
    text = re.sub(r"\s+", " ", text).strip()
    meaningless = {"∆°","·ªù","·ªùm","hmm","ggg"}
    tokens = [t for t in text.split() if t not in meaningless]
    return " ".join(tokens)
df['clean_text'] = df['review_text'].apply(preprocess_text)
df['label'] = df['rating'].apply(lambda r: 2 if r>=4 else 1 if r==3 else 0)
df.to_csv("itemReview_pxquy_preprocessed.csv", index=False)
print("Preprocessed saved ‚úî")

# --------------------------
# d) Tokenize + train/val split
# --------------------------
SAMPLE_SIZE = min(20000, len(df))
df_small = df.sample(SAMPLE_SIZE, random_state=SEED).reset_index(drop=True)
MAX_WORDS = 20000
MAX_LEN = 150
tokenizer = Tokenizer(num_words=MAX_WORDS, oov_token="<UNK>")
tokenizer.fit_on_texts(df_small['clean_text'])
X = pad_sequences(tokenizer.texts_to_sequences(df_small['clean_text']), maxlen=MAX_LEN)
y = to_categorical(df_small['label'], num_classes=3)

X_train, X_val, y_train, y_val = train_test_split(X,y,test_size=0.2,random_state=SEED,stratify=np.argmax(y,axis=1))
cw_vals = compute_class_weight('balanced', classes=np.unique(np.argmax(y_train,1)), y=np.argmax(y_train,1))
class_weight = {i:float(w) for i,w in enumerate(cw_vals)}

# --------------------------
# e) Build models
# --------------------------
EMBED_DIM = 128
def build_cnn():
    m=Sequential([
        Embedding(MAX_WORDS,EMBED_DIM,input_length=MAX_LEN),
        Conv1D(128,5,activation='relu',padding='same'),BatchNormalization(),
        Conv1D(64,5,activation='relu',padding='same'),BatchNormalization(),
        GlobalMaxPooling1D(),Dropout(0.5),
        Dense(64,activation='relu'),Dropout(0.4),
        Dense(3,activation='softmax')
    ]);m.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy']);return m

def build_rnn():
    m=Sequential([
        Embedding(MAX_WORDS,EMBED_DIM,input_length=MAX_LEN),
        SimpleRNN(128,return_sequences=True),Dropout(0.4),
        SimpleRNN(64),Dropout(0.4),
        Dense(64,activation='relu'),Dropout(0.3),
        Dense(3,activation='softmax')
    ]);m.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy']);return m

def build_lstm():
    m=Sequential([
        Embedding(MAX_WORDS,EMBED_DIM,input_length=MAX_LEN),
        Bidirectional(LSTM(64,return_sequences=True)),Dropout(0.4),
        Bidirectional(LSTM(32)),Dropout(0.4),
        Dense(64,activation='relu'),Dropout(0.3),
        Dense(3,activation='softmax')
    ]);m.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy']);return m

# --------------------------
# f) Train models
# --------------------------
EPOCHS=12;BATCH=64
es=EarlyStopping(monitor='val_loss',patience=3,restore_best_weights=True)
rlr=ReduceLROnPlateau(monitor='val_loss',factor=0.5,patience=2)

def train_model(name,builder):
    model=builder();ck=f"best_{name}_pxquy.h5"
    ckpt=ModelCheckpoint(ck,monitor='val_accuracy',save_best_only=True)
    print(f"Training {name}...")
    hist=model.fit(X_train,y_train,validation_data=(X_val,y_val),epochs=EPOCHS,
                   batch_size=BATCH,callbacks=[es,rlr,ckpt],class_weight=class_weight,verbose=1)
    best=tf.keras.models.load_model(ck,compile=False)
    yprob=best.predict(X_val,verbose=0)
    ypred=np.argmax(yprob,1);ytrue=np.argmax(y_val,1)
    acc=accuracy_score(ytrue,ypred);f1=f1_score(ytrue,ypred,average='macro')
    return {"acc":acc,"f1":f1,"history":hist,"model":best,"path":ck,
            "conf":confusion_matrix(ytrue,ypred)}

results={}
for nm,builder in [("CNN",build_cnn),("RNN",build_rnn),("LSTM",build_lstm)]:
    try: results[nm]=train_model(nm,builder)
    except Exception as e: print(f"‚ùå {nm} failed:",e)

# --------------------------
# g) Select best + save deploy assets
# --------------------------
best_name=max(results,key=lambda k:results[k]['acc'])
best_path=results[best_name]['path']
deploy_dir=Path("deploy_assets");deploy_dir.mkdir(exist_ok=True)
with open(deploy_dir/"tokenizer_pxquy.pkl","wb") as f: pickle.dump(tokenizer,f)
with open(deploy_dir/"training_config.json","w",encoding="utf-8") as f:
    json.dump({"MAX_LEN":MAX_LEN,"MAX_WORDS":MAX_WORDS,"best_model":best_path},f,ensure_ascii=False,indent=2)
label_map={0:"negative",1:"neutral",2:"positive"}
with open(deploy_dir/"label_map.json","w",encoding="utf-8") as f:
    json.dump(label_map,f,ensure_ascii=False,indent=2)
shutil.copy(best_path,f"best_model_pxquy_{best_name}.h5")
print(f"‚úÖ Best model: {best_name}, acc={results[best_name]['acc']:.4f}, f1={results[best_name]['f1']:.4f}")

# --------------------------
# h) Visualization
# --------------------------
names=list(results.keys())
accs=[results[n]['acc'] for n in names]
f1s=[results[n]['f1'] for n in names]

plt.figure(figsize=(8,4))
plt.subplot(1,2,1)
plt.bar(names,accs,color="skyblue");plt.title("Validation Accuracy");plt.ylim(0,1)
plt.subplot(1,2,2)
plt.bar(names,f1s,color="lightgreen");plt.title("Validation F1-score");plt.ylim(0,1)
plt.tight_layout();plt.savefig("pxquy_model_metrics.png");plt.show()

for n in names:
    conf=results[n]['conf']
    plt.figure(figsize=(4,3))
    sns.heatmap(conf,annot=True,fmt='d',cmap='Blues',
                xticklabels=label_map.values(),yticklabels=label_map.values())
    plt.title(f"Confusion Matrix - {n}")
    plt.xlabel("Predicted");plt.ylabel("True")
    plt.tight_layout();plt.show()
# --------------------------
# üìà Bi·ªÉu ƒë·ªì Loss & Accuracy theo epoch cho t·ª´ng m√¥ h√¨nh
# --------------------------
def plot_history(history, model_name):
    hist = history.history
    epochs = range(1, len(hist['loss']) + 1)

    plt.figure(figsize=(10,4))
    # Loss
    plt.subplot(1,2,1)
    plt.plot(epochs, hist['loss'], 'b-', label='Train Loss')
    plt.plot(epochs, hist['val_loss'], 'r--', label='Val Loss')
    plt.title(f'{model_name} - Loss')
    plt.xlabel('Epoch'); plt.ylabel('Loss')
    plt.legend()

    # Accuracy
    plt.subplot(1,2,2)
    plt.plot(epochs, hist['accuracy'], 'b-', label='Train Acc')
    plt.plot(epochs, hist['val_accuracy'], 'r--', label='Val Acc')
    plt.title(f'{model_name} - Accuracy')
    plt.xlabel('Epoch'); plt.ylabel('Accuracy')
    plt.legend()

    plt.suptitle(f'Training Curves - {model_name}')
    plt.tight_layout()
    plt.savefig(f"{model_name}_pxquy_history.png")
    plt.show()

# V·∫Ω cho t·ª´ng model c√≥ trong results
for model_name, info in results.items():
    print(f"\nüìä Bi·ªÉu ƒë·ªì loss/accuracy cho {model_name}")
    plot_history(info['history'], model_name)

# --------------------------
# Sanity check
# --------------------------
best_model=tf.keras.models.load_model(f"best_model_pxquy_{best_name}.h5",compile=False)
samples=[
    "S·∫£n ph·∫©m qu√° t·ªá, m√¨nh r·∫•t th·∫•t v·ªçng",
    "Kh√¥ng t·ªët ch√∫t n√†o",
    "Mua v·ªÅ r·∫•t h√†i l√≤ng, ƒë√°ng ti·ªÅn",
    "M·∫∑c d√π ƒë·∫πp nh∆∞ng ch·∫•t l∆∞·ª£ng k√©m"
]
s_clean=[preprocess_text(s) for s in samples]
s_seq=tokenizer.texts_to_sequences(s_clean)
s_X=pad_sequences(s_seq,maxlen=MAX_LEN)
probs=best_model.predict(s_X,verbose=0)
for s,p in zip(samples,probs):
    pred=np.argmax(p);print(f"{s} -> {label_map[pred]} (prob={p[pred]:.3f})")

print("\n=== ‚úÖ Pipeline completed. Assets saved in deploy_assets/ ===")
```
