```python
# Nguyễn Minh Vũ - B22DCVT594
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import (Input, Conv2D, MaxPooling2D, Flatten, Dense, Dropout,
                                     BatchNormalization, Activation, GlobalAveragePooling2D)
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint

plt.style.use('ggplot')
tf.random.set_seed(42)
np.random.seed(42)

# -------------------------
# 1. Chuẩn bị dữ liệu
# -------------------------
print("1) Tải và chuẩn bị dữ liệu MNIST...")
(X_train_raw, y_train), (X_test_raw, y_test) = mnist.load_data()

# Tạo 100 ảnh test tự tạo (giả lập ảnh thực tế) từ bộ test gốc
indices = np.random.choice(len(X_test_raw), 100, replace=False)
X_temp = X_test_raw[indices]
y_custom = y_test[indices]

# Augment + noise cho 100 ảnh test (giữ làm tập kiểm tra thực tế)
test_datagen = ImageDataGenerator(rotation_range=25, zoom_range=0.18,
                                  width_shift_range=0.12, height_shift_range=0.12,
                                  fill_mode='nearest')
X_custom_aug = []
for img in X_temp:
    img = np.expand_dims(img, -1)
    img_aug = test_datagen.random_transform(img)
    noise = np.random.normal(0, 12, img_aug.shape)  # thêm gaussian noise
    img_final = np.clip(img_aug + noise, 0, 255)
    X_custom_aug.append(img_final)
X_custom = np.array(X_custom_aug).reshape(-1, 28, 28, 1)

# Chuẩn hoá toàn bộ dữ liệu (train + custom test)
X_train = X_train_raw.astype('float32') / 255.0
X_train = np.expand_dims(X_train, -1)
X_custom = X_custom.astype('float32') / 255.0

y_train_hot = to_categorical(y_train, 10)
y_custom_hot = to_categorical(y_custom, 10)

print(f"  - Train shape: {X_train.shape}, Train labels: {y_train_hot.shape}")
print(f"  - Custom test shape: {X_custom.shape}, Custom labels: {y_custom_hot.shape}")

# Vẽ vài ảnh mẫu và phân bố (tuỳ chọn)
plt.figure(figsize=(12,4))
plt.subplot(1,3,1)
sns.countplot(y_train)
plt.title("Phân bố nhãn Train (MNIST)")

plt.subplot(1,3,2)
sns.countplot(y_custom)
plt.title("Phân bố nhãn Custom Test (100 ảnh)")

plt.subplot(1,3,3)
plt.imshow(X_custom[0].reshape(28,28), cmap='gray')
plt.title(f"Ví dụ Custom Test (label={y_custom[0]})")
plt.axis('off')
plt.tight_layout()
plt.show()

# -------------------------
# 2. Data Augmentation cho TRAIN
# -------------------------
train_datagen = ImageDataGenerator(
    rotation_range=15,
    width_shift_range=0.10,
    height_shift_range=0.10,
    zoom_range=0.12,
    shear_range=0.06,
    horizontal_flip=False,
    fill_mode='nearest'
)

train_generator = train_datagen.flow(X_train, y_train_hot, batch_size=128, shuffle=True)

# -------------------------
# 3. Xây dựng mô hình Keras cải tiến
#    - BatchNormalization
#    - Dropout ở nhiều chỗ
#    - GlobalAveragePooling (nhẹ)
# -------------------------
def build_improved_model(input_shape=(28,28,1), num_classes=10):
    model = Sequential(name="Improved_CNN_MNIST")
    model.add(Input(shape=input_shape))

    # Block 1
    model.add(Conv2D(32, (3,3), padding='same'))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(Conv2D(32, (3,3), padding='same'))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(MaxPooling2D((2,2)))
    model.add(Dropout(0.25))

    # Block 2
    model.add(Conv2D(64, (3,3), padding='same'))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(Conv2D(64, (3,3), padding='same'))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(MaxPooling2D((2,2)))
    model.add(Dropout(0.30))

    # Optional deeper block
    model.add(Conv2D(128, (3,3), padding='same'))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(GlobalAveragePooling2D())
    model.add(Dropout(0.45))

    # Output
    model.add(Dense(num_classes, activation='softmax'))
    return model

model = build_improved_model()
model.summary()

# -------------------------
# 4. Compile + callbacks
# -------------------------
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
              loss='categorical_crossentropy',
              metrics=['accuracy'])

callbacks = [
    EarlyStopping(monitor='val_loss', patience=4, restore_best_weights=True, verbose=1),
    ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=2, min_lr=1e-6, verbose=1),
    ModelCheckpoint("best_model_mnist.h5", monitor='val_loss', save_best_only=True, verbose=1)
]

# -------------------------
# 5. Train
# -------------------------
EPOCHS = 15
STEPS_PER_EPOCH = X_train.shape[0] // 128

history = model.fit(
    train_generator,
    steps_per_epoch=STEPS_PER_EPOCH,
    epochs=EPOCHS,
    validation_split=0.0,  # chúng ta dùng validation bằng tay từ flow -> dùng validation_data instead
    validation_data=(X_train[:6000], y_train_hot[:6000]),  # dùng 6000 sạch làm validation (tùy chọn)
    callbacks=callbacks,
    verbose=2
)

# -------------------------
# 6. Đánh giá trên tập custom (100 ảnh sinh viên tạo)
# -------------------------
print("\nĐánh giá trên 100 ảnh tự tạo (X_custom):")
eval_custom = model.evaluate(X_custom, y_custom_hot, verbose=0)
print(f"  - Loss (custom): {eval_custom[0]:.4f} | Accuracy (custom): {eval_custom[1]*100:.2f}%")

# Đánh giá trên tập train nhỏ (kiểm tra overfit)
eval_train_small = model.evaluate(X_train[:5000], y_train_hot[:5000], verbose=0)
print(f"  - Loss (train 5k): {eval_train_small[0]:.4f} | Accuracy (train 5k): {eval_train_small[1]*100:.2f}%")

# -------------------------
# 7. Vẽ đồ thị Loss & Accuracy
# -------------------------
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.plot(history.history['loss'], label='train_loss', marker='o')
plt.plot(history.history['val_loss'], label='val_loss', marker='o')
plt.title('Loss theo epoch')
plt.xlabel('Epoch'); plt.ylabel('Loss')
plt.legend()

plt.subplot(1,2,2)
plt.plot(history.history['accuracy'], label='train_acc', marker='o')
plt.plot(history.history['val_accuracy'], label='val_acc', marker='o')
plt.title('Accuracy theo epoch')
plt.xlabel('Epoch'); plt.ylabel('Accuracy')
plt.legend()
plt.tight_layout()
plt.show()

# -------------------------
# 8. Minh họa dự đoán trên 10 ảnh custom
# -------------------------
preds = np.argmax(model.predict(X_custom, verbose=0), axis=1)
plt.figure(figsize=(12,2.5))
for i in range(10):
    plt.subplot(1,10,i+1)
    plt.imshow(X_custom[i].reshape(28,28), cmap='gray')
    color = 'green' if preds[i] == y_custom[i] else 'red'
    plt.title(f"P:{preds[i]}", color=color)
    plt.axis('off')
plt.suptitle("Dự đoán trên ảnh tự tạo (Xanh: đúng, Đỏ: sai)")
plt.show()

# -------------------------
# 9. In báo cáo ngắn (tóm tắt)
# -------------------------
print("\n--- Báo cáo tóm tắt ---")
print(f"Accuracy trên 100 ảnh tự tạo: {eval_custom[1]*100:.2f}%")
print("Cải tiến đã áp dụng: Train-time augmentation, BatchNormalization, Dropout nhiều vị trí, EarlyStopping, ReduceLROnPlateau.")
print("Gợi ý tiếp theo: nếu muốn tăng robustness hơn nữa, thử tăng augmentation (random erasing), hoặc huấn luyện ensemble / dùng kiến trúc pretrained nhẹ như MobileNetV2 (transfer learning).")
```
