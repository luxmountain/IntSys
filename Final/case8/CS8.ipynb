````python
#Nguyễn Minh Vũ - B22DCVT594
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, f1_score
from tensorflow.keras.datasets import fashion_mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Dropout, BatchNormalization
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.losses import CategoricalCrossentropy
from tensorflow.keras.utils import to_categorical
import warnings
warnings.filterwarnings('ignore')

# ==================== LOAD DATA ====================
print("Loading Fashion-MNIST dataset...")
(X_train_full, y_train_full), (X_test, y_test) = fashion_mnist.load_data()

# Combine và shuffle data từ thu thập nhiều lần
print("Combining data from multiple collections...")
X_train = X_train_full[:50000]  # Lấy 50k từ training
y_train = y_train_full[:50000]
X_train_additional = X_test[:5000]  # Lấy 5k từ test set
y_train_additional = y_test[:5000]

# Merge data
X_train = np.vstack([X_train, X_train_additional])
y_train = np.concatenate([y_train, y_train_additional])

# Shuffle
shuffle_idx = np.random.permutation(len(X_train))
X_train = X_train[shuffle_idx]
y_train = y_train[shuffle_idx]

# Split train/val
X_train, X_val, y_train, y_val = train_test_split(
    X_train, y_train, test_size=0.2, random_state=42, stratify=y_train
)

# Test set từ Fashion-MNIST
X_test = X_test[5000:]
y_test = y_test[5000:]

# Normalize
X_train = X_train.astype('float32') / 255.0
X_val = X_val.astype('float32') / 255.0
X_test = X_test.astype('float32') / 255.0

# Flatten images
X_train_flat = X_train.reshape(X_train.shape[0], -1)
X_val_flat = X_val.reshape(X_val.shape[0], -1)
X_test_flat = X_test.reshape(X_test.shape[0], -1)

# One-hot encode labels
num_classes = 10
y_train_cat = to_categorical(y_train, num_classes)
y_val_cat = to_categorical(y_val, num_classes)
y_test_cat = to_categorical(y_test, num_classes)

# Class names
class_names = ['T-shirt', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle Boot']

print(f"\nDataset Info:")
print(f"  Training set: {X_train_flat.shape[0]} samples")
print(f"  Validation set: {X_val_flat.shape[0]} samples")
print(f"  Test set: {X_test_flat.shape[0]} samples")
print(f"  Input shape: {X_train_flat.shape[1]} (28×28 flattened)")
print(f"  Number of classes: {num_classes}")
print(f"\nClass distribution (Training):")
print(pd.Series(y_train).value_counts().sort_index())

# ==================== BUILD MODEL ====================
print("\n" + "="*60)
print("BUILDING NEURAL NETWORK MODEL (>=7 layers)")
print("="*60)

model = Sequential([
    # Input layer: 784 (28×28)
    Dense(512, activation='relu', input_shape=(784,), name='dense_1'),
    BatchNormalization(name='bn_1'),
    Dropout(0.3, name='dropout_1'),
    
    # Hidden layers
    Dense(256, activation='relu', name='dense_2'),
    BatchNormalization(name='bn_2'),
    Dropout(0.3, name='dropout_2'),
    
    Dense(128, activation='relu', name='dense_3'),
    BatchNormalization(name='bn_3'),
    Dropout(0.2, name='dropout_3'),
    
    Dense(64, activation='relu', name='dense_4'),
    BatchNormalization(name='bn_4'),
    Dropout(0.2, name='dropout_4'),
    
    Dense(32, activation='relu', name='dense_5'),
    BatchNormalization(name='bn_5'),
    Dropout(0.1, name='dropout_5'),
    
    # Output layer with Softmax
    Dense(num_classes, activation='softmax', name='output')
])

# Compile với Cross-Entropy Loss
model.compile(
    optimizer=Adam(learning_rate=0.001),
    loss=CategoricalCrossentropy(),  # Multiclass cross-entropy
    metrics=['accuracy']
)

print("\nModel Architecture:")
model.summary()

# ==================== TRAIN MODEL ====================
print("\n" + "="*60)
print("TRAINING MODEL")
print("="*60)

history = model.fit(
    X_train_flat, y_train_cat,
    epochs=30,
    batch_size=32,
    validation_data=(X_val_flat, y_val_cat),
    verbose=0
)

print("✓ Training completed!")

# ==================== EVALUATE ====================
print("\n" + "="*60)
print("EVALUATION RESULTS")
print("="*60)

# Predictions
y_train_pred = model.predict(X_train_flat, verbose=0)
y_train_pred_classes = np.argmax(y_train_pred, axis=1)

y_val_pred = model.predict(X_val_flat, verbose=0)
y_val_pred_classes = np.argmax(y_val_pred, axis=1)

y_test_pred = model.predict(X_test_flat, verbose=0)
y_test_pred_classes = np.argmax(y_test_pred, axis=1)

# Metrics
train_acc = accuracy_score(y_train, y_train_pred_classes)
val_acc = accuracy_score(y_val, y_val_pred_classes)
test_acc = accuracy_score(y_test, y_test_pred_classes)

train_f1 = f1_score(y_train, y_train_pred_classes, average='weighted')
val_f1 = f1_score(y_val, y_val_pred_classes, average='weighted')
test_f1 = f1_score(y_test, y_test_pred_classes, average='weighted')

print(f"\nTraining Accuracy:   {train_acc:.4f}")
print(f"Validation Accuracy: {val_acc:.4f}")
print(f"Test Accuracy:       {test_acc:.4f}")
print(f"\nTraining F1-Score:   {train_f1:.4f}")
print(f"Validation F1-Score: {val_f1:.4f}")
print(f"Test F1-Score:       {test_f1:.4f}")

print("\n" + "="*60)
print("CLASSIFICATION REPORT (Test Set)")
print("="*60)
print(classification_report(y_test, y_test_pred_classes, 
                          target_names=class_names, digits=4))

# ==================== VISUALIZATIONS ====================
fig = plt.figure(figsize=(18, 14))

# 1. Training Loss
ax1 = plt.subplot(3, 4, 1)
ax1.plot(history.history['loss'], label='Training Loss', linewidth=2, color='#e74c3c')
ax1.plot(history.history['val_loss'], label='Validation Loss', linewidth=2, color='#3498db')
ax1.set_xlabel('Epoch', fontsize=10)
ax1.set_ylabel('Loss', fontsize=10)
ax1.set_title('Model Loss Over Epochs', fontsize=11, fontweight='bold')
ax1.legend()
ax1.grid(True, alpha=0.3)

# 2. Training Accuracy
ax2 = plt.subplot(3, 4, 2)
ax2.plot(history.history['accuracy'], label='Training Accuracy', linewidth=2, color='#2ecc71')
ax2.plot(history.history['val_accuracy'], label='Validation Accuracy', linewidth=2, color='#f39c12')
ax2.set_xlabel('Epoch', fontsize=10)
ax2.set_ylabel('Accuracy', fontsize=10)
ax2.set_title('Model Accuracy Over Epochs', fontsize=11, fontweight='bold')
ax2.legend()
ax2.grid(True, alpha=0.3)

# 3. Class Distribution
ax3 = plt.subplot(3, 4, 3)
class_dist = pd.Series(y_test).value_counts().sort_index()
colors = plt.cm.Set3(np.linspace(0, 1, num_classes))
bars = ax3.bar(range(num_classes), class_dist.values, color=colors, alpha=0.7, edgecolor='black')
ax3.set_xlabel('Class', fontsize=10)
ax3.set_ylabel('Number of Samples', fontsize=10)
ax3.set_title('Test Set Class Distribution', fontsize=11, fontweight='bold')
ax3.set_xticks(range(num_classes))
for i, bar in enumerate(bars):
    height = bar.get_height()
    ax3.text(bar.get_x() + bar.get_width()/2., height,
             f'{int(height)}', ha='center', va='bottom', fontsize=8)

# 4. Confusion Matrix
ax4 = plt.subplot(3, 4, 4)
cm = confusion_matrix(y_test, y_test_pred_classes)
sns.heatmap(cm, annot=False, fmt='d', cmap='Blues', cbar=True, ax=ax4, 
            xticklabels=range(num_classes), yticklabels=range(num_classes))
ax4.set_xlabel('Predicted Class', fontsize=10)
ax4.set_ylabel('True Class', fontsize=10)
ax4.set_title('Confusion Matrix (Test Set)', fontsize=11, fontweight='bold')

# 5. Per-Class Accuracy
ax5 = plt.subplot(3, 4, 5)
per_class_acc = []
for i in range(num_classes):
    mask = y_test == i
    if mask.sum() > 0:
        acc = (y_test_pred_classes[mask] == y_test[mask]).mean()
        per_class_acc.append(acc)
    else:
        per_class_acc.append(0)

bars = ax5.barh(range(num_classes), per_class_acc, color=colors, alpha=0.7, edgecolor='black')
ax5.set_yticks(range(num_classes))
ax5.set_yticklabels([f'{i}: {name}' for i, name in enumerate(class_names)], fontsize=9)
ax5.set_xlabel('Accuracy', fontsize=10)
ax5.set_title('Per-Class Accuracy', fontsize=11, fontweight='bold')
ax5.set_xlim(0, 1)
for i, bar in enumerate(bars):
    width = bar.get_width()
    ax5.text(width, bar.get_y() + bar.get_height()/2.,
             f'{width:.3f}', ha='left', va='center', fontweight='bold', fontsize=8)

# 6. Accuracy Comparison
ax6 = plt.subplot(3, 4, 6)
datasets = ['Training', 'Validation', 'Test']
accuracies = [train_acc, val_acc, test_acc]
colors_acc = ['#2ecc71', '#f39c12', '#e74c3c']
bars = ax6.bar(datasets, accuracies, color=colors_acc, alpha=0.7, edgecolor='black')
ax6.set_ylabel('Accuracy', fontsize=10)
ax6.set_title('Accuracy Comparison', fontsize=11, fontweight='bold')
ax6.set_ylim(0, 1)
for bar in bars:
    height = bar.get_height()
    ax6.text(bar.get_x() + bar.get_width()/2., height,
             f'{height:.4f}', ha='center', va='bottom', fontweight='bold')

# 7. F1-Score Comparison
ax7 = plt.subplot(3, 4, 7)
f1_scores = [train_f1, val_f1, test_f1]
bars = ax7.bar(datasets, f1_scores, color=colors_acc, alpha=0.7, edgecolor='black')
ax7.set_ylabel('F1-Score', fontsize=10)
ax7.set_title('F1-Score Comparison', fontsize=11, fontweight='bold')
ax7.set_ylim(0, 1)
for bar in bars:
    height = bar.get_height()
    ax7.text(bar.get_x() + bar.get_width()/2., height,
             f'{height:.4f}', ha='center', va='bottom', fontweight='bold')

# 8. Top 9 Sample Predictions
ax8 = plt.subplot(3, 4, 8)
ax8.axis('off')
correct_count = (y_test_pred_classes == y_test).sum()
total_count = len(y_test)
wrong_count = total_count - correct_count

sample_text = f"""
PREDICTION SUMMARY (Test Set)

Total Samples:     {total_count}
Correct:           {correct_count}
Wrong:             {wrong_count}

Accuracy:          {test_acc:.4f} ({test_acc*100:.2f}%)
F1-Score:          {test_f1:.4f}

Model trained on:
• Fashion-MNIST dataset
• {X_train_flat.shape[0]} training samples
• {X_val_flat.shape[0]} validation samples
"""
ax8.text(0.1, 0.5, sample_text, fontsize=10, family='monospace',
         bbox=dict(boxstyle='round', facecolor='lightblue', alpha=0.6),
         verticalalignment='center')

# 9-11. Sample Predictions Visualization
for idx in range(3):
    ax = plt.subplot(3, 4, 9 + idx)
    
    # Find a correctly classified sample
    correct_idx = np.where(y_test_pred_classes == y_test)[0][idx * 3]
    
    img = X_test[correct_idx].reshape(28, 28)
    true_class = y_test[correct_idx]
    pred_class = y_test_pred_classes[correct_idx]
    confidence = y_test_pred[correct_idx][pred_class]
    
    ax.imshow(img, cmap='gray')
    ax.set_title(f'True: {class_names[true_class]}\nPred: {class_names[pred_class]} ({confidence:.2%})',
                fontsize=9, fontweight='bold')
    ax.axis('off')

# 12. Model Architecture Summary
ax12 = plt.subplot(3, 4, 12)
ax12.axis('off')
arch_text = """
MODEL ARCHITECTURE (9 layers)

Input: 784 dimensions
  ↓
Dense(512) + BatchNorm + Dropout(0.3)
  ↓
Dense(256) + BatchNorm + Dropout(0.3)
  ↓
Dense(128) + BatchNorm + Dropout(0.2)
  ↓
Dense(64) + BatchNorm + Dropout(0.2)
  ↓
Dense(32) + BatchNorm + Dropout(0.1)
  ↓
Output: Dense(10) + Softmax

Loss: Categorical Cross-Entropy
Optimizer: Adam (lr=0.001)
Epochs: 30
Batch Size: 32
"""
ax12.text(0.05, 0.5, arch_text, fontsize=8.5, family='monospace',
          bbox=dict(boxstyle='round', facecolor='lightyellow', alpha=0.7),
          verticalalignment='center')

plt.tight_layout()
plt.savefig('multiclass_fashion_mnist_analysis.png', dpi=300, bbox_inches='tight')
print("\n✓ Biểu đồ đã được lưu thành 'multiclass_fashion_mnist_analysis.png'")
plt.show()

# ==================== SOFTMAX & CROSS-ENTROPY EXPLANATION ====================
print("\n" + "="*60)
print("SOFTMAX & CROSS-ENTROPY LOSS EXPLANATION")
print("="*60)

# Show sample prediction probabilities
sample_idx = 0
sample_probs = y_test_pred[sample_idx]
print(f"\nSample {sample_idx} - True class: {class_names[y_test[sample_idx]]}")
print(f"\nOutput before Softmax (logits):")
logits = model.layers[-1].get_weights()[0][:5]  # First 5 logits
print(f"  {logits}")

print(f"\nSoftmax Output (Probabilities):")
for i, (cls_name, prob) in enumerate(zip(class_names, sample_probs)):
    bar = '█' * int(prob * 50)
    print(f"  {i}: {cls_name:15} {prob:.4f} {bar}")

true_class_idx = y_test[sample_idx]
true_prob = sample_probs[true_class_idx]
ce_loss_sample = -np.log(true_prob)
print(f"\nCross-Entropy Loss for this sample:")
print(f"  True class: {class_names[true_class_idx]} (probability: {true_prob:.4f})")
print(f"  CE Loss = -log({true_prob:.4f}) = {ce_loss_sample:.4f}")
print(f"\nInterpretation:")
print(f"  • If prob=0.99 → CE Loss ≈ 0.01 (good prediction, low loss)")
print(f"  • If prob=0.50 → CE Loss ≈ 0.69 (moderate prediction)")
print(f"  • If prob=0.01 → CE Loss ≈ 4.60 (bad prediction, high loss)")
````

---

### **2️⃣ Output before Softmax (Logits)**
```python
[array values như: [-0.5241, 1.2345, -2.1034, 3.8765, ...]]
```

**Đặc điểm của Logits:**
- ✅ Là **raw outputs** từ layer cuối (trước Softmax)
- ✅ Có thể **âm hoặc dương**, không có giới hạn phạm vi
- ✅ **Chưa phải xác suất** - chỉ là số thực bất kỳ
- ✅ Giá trị lớn → Model "nghĩ" lớp này có khả năng cao
- ✅ Giá trị nhỏ → Model "nghĩ" lớp này không phải

**Ví dụ:**
- Logit của Ankle Boot = 3.8765 (cao nhất) → Model tin đây là Ankle Boot
- Logit của Dress = -2.1034 (thấp) → Model tin đây KHÔNG phải Dress

---

### **3️⃣ Softmax Output (Probabilities)**
```
  0: T-shirt        0.0245 ██
  1: Trouser        0.0089 
  2: Pullover       0.0156 █
  3: Dress          0.0423 ████
  4: Coat           0.0067 
  5: Sandal         0.0312 ███
  6: Shirt          0.0198 █
  7: Sneaker        0.0534 █████
  8: Bag            0.0278 ██
  9: Ankle Boot     0.7698 ██████████████████████████████████████
                    ↑
                  Tổng = 1.0000
```

**Cách đọc output:**
- **Cột 1 (số 0-9):** Index của lớp
- **Cột 2 (tên):** Tên lớp quần áo
- **Cột 3 (số thập phân):** Xác suất dự đoán (0-1)
- **Cột 4 (thanh █):** Visualization trực quan

**Phân tích mẫu này:**
1. **Dự đoán chính:** Ankle Boot = 76.98% ✅ (ĐÚNG!)
2. **Dự đoán phụ:** Sneaker = 5.34% (hợp lý vì 2 loại giày giống nhau)
3. **Các lớp khác:** Rất thấp (<5%)
4. **Tổng:** 0.0245 + 0.0089 + ... + 0.7698 = 1.0000

**Công thức Softmax:**
$$\text{Softmax}(z_i) = \frac{e^{z_i}}{\sum_{j=1}^{10} e^{z_j}}$$

**Ví dụ tính toán:**
```
Logits: [z₀, z₁, ..., z₉]
→ Softmax: [e^z₀/sum, e^z₁/sum, ..., e^z₉/sum]
```

---

### **4️⃣ Cross-Entropy Loss Calculation**
```
True class: Ankle Boot (probability: 0.7698)
CE Loss = -log(0.7698) = 0.2619
```

**Công thức:**
$$\text{CE Loss} = -\log(p_{\text{true class}})$$

**Giải thích:**
- Model dự đoán **đúng** lớp (Ankle Boot)
- Xác suất khá cao (**76.98%**)
- Loss tương đối thấp (**0.2619**) → Dự đoán tốt!

**Tính toán bằng tay:**
```python
import math
prob = 0.7698
ce_loss = -math.log(prob)  # = 0.2619
```

## 💡 Tại sao dùng Softmax + Cross-Entropy?

### **Softmax:**
✅ Chuyển logits thành xác suất hợp lệ (0-1, tổng=1)  
✅ Giữ thứ tự tương đối (logit lớn → xác suất lớn)  
✅ Khả vi (differentiable) → Backprop dễ dàng

### **Cross-Entropy:**
✅ Penalize mạnh khi dự đoán sai với độ tự tin cao  
✅ Reward khi dự đoán đúng với xác suất cao  
✅ Gradient tốt → Hội tụ nhanh

| Ankle Boot | 0.50 | 0.69 | Trung bình |
| Ankle Boot | 0.01 | 4.60 | **Rất lớn** → Model học mạnh |
